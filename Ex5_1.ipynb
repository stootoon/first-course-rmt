# Ex 5.1.2
## (a)
$$P(\lambda_1, \lambda_2) = |\lambda_1 - \lambda_2| \exp(-{N \over 4 \sigma^2}(\lambda_1^2 + \lambda_2^2)).$$

## (b)
Switching to $$ \begin{bmatrix} \lambda_+ \\ \lambda_- \end{bmatrix} = \begin{bmatrix} 1 & +1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} \lambda_1 \\ \lambda_2 \end{bmatrix},$$ so the absolute value of the determinant of the Jacobian is 2 (though we won't use this since this is a constant and $P$ is unnormalized). 
We also have $$ \begin{align*}
\lambda_1 &= (\lambda_+ - \lambda_-)/2\\
\lambda_2 &= (\lambda_+ + \lambda_-)/2\\
\lambda_1^2 + \lambda_2^2 &= {1 \over 4}(\lambda_+^2 + \lambda_-^2 - 2\lambda_+ \lambda_- + \lambda_+^2 + \lambda_-^2 + 2\lambda_+ \lambda_- )\\
&= {1 \over 2}(\lambda_+^2 + \lambda_-^2)
\end{align*}$$
Combining these, we get $$P(\lambda_+, \lambda_-)\propto {1\over 2 } |\lambda_-| \exp(-{1 \over 4 \sigma^2}(\lambda_+^2 + \lambda_-^2))$$
$\lambda_+$ is isolated and integrates out, so we get $$ P(\lambda_-) \propto |\lambda_-| \exp(-{1 \over 4 \sigma^2}\lambda_-^2),$$ or in terms of $x = |\lambda_-|$, $$P(x) \propto x \; e^{-x^2/4 \sigma^2}, \quad x \ge 0$$  

## (c)
$$ \int_0^\infty x e^{-x^2/4\sigma^2} dx = -2 \sigma^2 \int -{x \over 2 \sigma^2} e^{-x^2/4\sigma^2} dx = 2 \sigma^2 \left.e^{-x^2/4\sigma^2}\right|_\infty^0 = 2 \sigma^2,$$ 
so 
$$ P(x) = {x \over 2\sigma^2} e^{-x^2/4\sigma^2} . $$

Then $$\begin{align*} E(x) &= \int_0^\infty {x^2 \over 2\sigma^2} e^{-x^2/4\sigma^2} dx \\ 
&= {1 \over 2} \int_{-\infty}^\infty {x^2 \over 2\sigma^2} e^{-x^2/4\sigma^2} dx\\
&= {1 \over 4 \sigma^2} \sqrt{4\pi \sigma^2 }\int_{-\infty}^\infty {1 \over \sqrt{4 \pi \sigma^2}} x^2 e^{-x^2/4\sigma^2} dx\\
&= {1 \over 4 \sigma^2} \sqrt{4\pi \sigma^2 } 2\sigma^2\\
&= \sqrt{\pi }\sigma.  
\end{align*} $$
Setting $\sigma^2 = \pi^{-1}$ makes the expectation 1. Then $$ P(x) = {\pi \over 2}x e^{-\pi x^2 \over 4}.$$
