# Ex 5.2.1

## (a)
$\lambda_2$ should equal 0, since e.g. the trace of the matrix should be 0 on average.

## (b)

$$V'(\lambda_2 = 0) = \lambda_2 = 0 = {2 \over 3}\left({1 \over 0 - 1} + {1 \over 0 + 1}\right)$$
so $\lambda_2 = 0$ is a solution.

## (c)
$$ g_3(z) = {1 \over 3}\left({1 \over z - 1} + {1 \over z} + {1 \over z + 1}\right) = {1 \over 3}{3 z^2 - 1 \over z^3 - z}.$$

## (d)
$$ \Pi_3(z) = {1 \over 3} \sum_{i=1}^3 {z - \lambda_i \over z - \lambda_i} = 1.$$

## (e)
$$g'(z) = {2z \over z^3 - z} - {{z^2 - {1 \over 3}} \over (z^3 - z)^2} (3 z^2 - 1) = {2 z \over z^3 - z} - 3 g_3 (z)^2 = {2 \over z^2 - 1} - 3 g_3(z)^2.$$ 
Left-hand-side:
$$ V'(z) g_3(z) - \Pi_3(z) = z g_3(z) - 1 = {z^2 - 1/3 \over z^2 - 1} -1 = {2 \over 3} {1 \over z^2 - 1}.$$
Right-hand-side
$$ {g'(z) \over 3} = {2 \over 3}{1 \over z^2 -1} - g_3(z)^2.$$
$$ {g'(z) \over 3 + g_3(z)^2} = {2 \over 3}{1 \over z^2 -1 }.$$
