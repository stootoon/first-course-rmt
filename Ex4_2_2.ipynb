## (a)
The important part of $g(z)$ is the $\sqrt{z^2 - 4}$. The branch cuts will be all the points where the argument to the square-root is negative real. That's because the squareroot will then send their angles to either $\pi/2$ or $-\pi/2$, depending on which side of the real axis we approached from, and will therefore contribute an $i$ or a $-i$ to $g(z)$. All points on the imaginary axis satisfy this condition, including $2i$.

## (b)
Below I'll just do the numerators.
$$ g_1(2i)= 2i - 2i \sqrt{1 + {4 \over 4 }} = 2i(1 - \sqrt{2}).$$

\begin{align*}
g_2(2i) &= 2i - \sqrt{2i - 2}\sqrt{2i+2}\\
&= 2i - \sqrt{\sqrt{8}e^{i3\pi/4}} \sqrt{\sqrt{8}e^{i\pi/4}} \\ &= 2i - 8^{1/4 }e^{i3 \pi/8} 8^{1/4}e^{i\pi/8}\\ &= 2i - \sqrt{8}e^{i\pi/2} \\ &= 2i - 2 \sqrt{2} i \\ &= 2i (1 - \sqrt{2}).
\end{align*}

Let $z = 2i \pm x$.
\begin{align*}
g_3(2i) &= z - \text{sign}(\Re(z))\sqrt{z^2 - 4}\\
&= 2i \mp \sqrt{-8 \pm 4 i x}\\
&= 2i \mp 2 \sqrt{-2 \pm i x}\\
&= 2i \mp 2\sqrt{2 e^{i \pm \pi}}\\
&= 2i \mp 2\sqrt{2} (\pm i)\\
&= 2i - 2i \sqrt{2}\\
&= 2i(1 - \sqrt{2}).
\end{align*}
