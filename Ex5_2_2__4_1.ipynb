## (b)
We can just integrate $\int_0^1 {1 \over x - \lambda} d\lambda$ because of the singularity at $x$, so we consider the principal value.

We split the integral around $x$. The left hand side is $$\int_0^{x - \epsilon} {1 \over x - \lambda} d\lambda =  -\left. \log(x - \lambda)\right|_0^{x - \epsilon} = \log(x) - \log(\epsilon).$$
The right hand side is $$\int_{x + \epsilon}^1 {1 \over x - \lambda} d\lambda =  -\left. \log(x - \lambda)\right|_{x + \epsilon}^1 = \log (-\epsilon) - \log (x - 1).$$
Adding the two pieces, we get $$ \log(x) - \log(\epsilon) + \log{-\epsilon \over x -1} = \log(x) - \log(\epsilon) + \log(\epsilon) - \log(1-x) =\log {x \over 1 - x}.$$
