## (a)

\begin{align*} 0 &= (z + q - 1)^2 - 4 qz \\ &= z^2 + 2 (q - 1)z + (q - 1)^2 - 4 qz \\ &= z^2 + 2 qz - 2 z - 4 q z + (q - 1)^2 \\ &= z^2 - 2 z ( q +1 ) + (q - 1)^2 \\\implies  z &= {2 (q + 1) \pm \sqrt{4 (q+1)^2 - 4 (q - 1)^2} \over 2} \\ &= (q + 1) \pm \sqrt{ (q+1)^2 - (q - 1)^2 }\\ &= (q + 1) \pm \sqrt{ 4 q}\\ &= q \pm 2 \sqrt{q} + 1 \\ &= (1 \pm \sqrt{q})^2. 
\end{align*} 

## (b)

The numerator is 
$$\begin{align*} z + q - 1 - \sqrt{(z - \lambda_-) (z - \lambda_+)} &= z + q - 1 -  \sqrt{z^2 - z((1 + \sqrt{q})(1 - \sqrt{q})+ [(1 + \sqrt{q})(1 - \sqrt{q})]^2)} \\ 
&= z + q - 1 -\sqrt{z^2 - 2 z(1 + q) + (1 - q)^2 }\\
&= z + q - 1 - z\sqrt{1 - z^{-1} 2(1 + q) + \dots}\\
&\to z + q - 1 -z(1 - z^{-1}(1+q))\\
&= z + q - 1 - z + 1 + q\\
&= 2 q\\
\implies g(z) &\to {2 q \over 2 q z } = {1 \over z}.
\end{align*}$$


### Moments

Let $\xi = z^{-1}$. Then
$$ 
\begin{align*}
g(z) &= \sum_{k=0}^\infty \xi^{k+1} \tau(A^k) = \xi \tau(A^0) + \xi^2 \tau(A^1) + \xi^3 \tau(A^2) + \dots
\end{align*}
$$
Then 
$$
\begin{align*} 
\left.{d g \over d \xi}\right|_{\xi=0}&= \tau(A^0) \\ 
\end{align*}
$$
Now $${d g \over d \xi} = { d g \over dz } {d z \over d \xi} ={ d g \over dz } \left({d \xi \over d z}\right)^{-1} = - g'(z) z^2.$$

$$ \boxed{\tau(A^0) = \lim_{z \to \infty} -g'(z) z^2 = 1.}$$

$$\left.{dg^2 \over d\xi^2}\right|_{\xi = 0} = 2 \tau(A^1).$$
$$ {dg^2 \over d\xi^2} = {d \over d \xi} {d g \over d \xi} = -{d \over d z} {dg \over d \xi} z^2 = (g''(z) z^2 + 2 g'(z) z) z^2 = g''(z) z^4 + 2 g'(z)z^3.$$
$$ \boxed{\tau(A^1) = {1 \over 2} \lim_{z \to \infty} (g''(z) z^4 + 2 g'(z)z^3) = {1 \over 2} 2 = 1.} $$
$$ \boxed{\tau(A^2) = {1 \over 6} \lim_{z \to \infty} {d ^3 g(z) \over d \xi^3 } = 1 + q.}$$ 

## (c)
### $q < 1$
When we analytically continue $g(z)$ from $z = \infty$ to 0 by following above the real line, each square root term contributes a factor of $+i$, so for $z$ near zero, $$g(z) = {z + q -1 + \sqrt{(\lambda_- - z) (\lambda_+ - z)} \over 2 q z} = {z + q -1 + \sqrt{(1 - q)^2 - 2(1 + q)z + z^2} \over 2 q z}.$$

For small $z$ we can approximate the squareroot as $$ \sqrt{(1 - q)^2 - 2(1 + q)z + z^2} \approx 1 - q - {(1 + q) \over 1 - q}z.$$

When we add this to the reset of the numerator, we get $$ g(z) = { (1 -q - 1 - q)z \over 2 q z (1 -q )} = -{1 \over 1 - q},$$ which is well behaved.  

### $\tau(\mathbf E^{-1})$
From Eqn 2.28, this is $g(0) = -{1 \over 1 -q}.$

### $q > 1$

Near $z = 0$, the analytic continuation is again $$ g(z) = {z + q - 1 + \sqrt{(\lambda_- - z)}\sqrt{(\lambda_+ - z)} \over 2 q z},$$ but we have to make sure we take out $(q-1)$ rather than $(1-q)$ at the relevant places, since sign changes etc. will be absorbed into the term preceding the square root.

Then $$g(z) \approx {z + q - 1 + \sqrt{\lambda_- \lambda_+ - z(\lambda_+ + \lambda_-)} \over 2 q z} \approx {z + q - 1 + q -1 - {q + 1 \over q -1} z \over 2 q z} = {2(q-1) - {2 \over q - 1}z \over 2 q z} \to {q -1 \over q z}.$$

Therefore there's a simple pole at 0 with value $(q-1)/q$.
